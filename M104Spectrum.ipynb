```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from scipy import constants as const
from astropy import units as u
frequency, flux = np.genfromtxt('photandseds.csv', delimiter=',', skip_header=2, usecols=(5,6), unpack=True)
plt.loglog(frequency, flux, markersize=2, color='blue', alpha=0.4, label='M104 Spectral Data') #The Messier 104 Galaxy was discovered in 1781 by Charles Messier's colleague, Pierre Méchain. M104's spirals are comprised of thick dust lanes, the site of star formation, and the galaxy is rich in globular clusters. The center of the galaxy is thought to be home to a massive black hole, and the galaxy is located in Virgo.
plt.title('M104 "Sombrero Galaxy" Flux Density vs. Frequency', size=13) #The IRAS data was taken in June of 1992 b the Einstein Observatory Imaging Instrument (IPC and HRI). The data includes 493 galaxis, specifically RSA and RC2 galaxies.
plt.xlabel('Frequency [1/Angstroms]', size=12) #The GALEX data was taken in January, 2011. This component of the Local Volume Legacy contains 390 galaxies within ~11Mpc of the Milky Way, 80% of which are dwarf galaxies.
plt.ylabel('Flux Density [ergs/s/cm^2]', size=12) #The IRAS data was taken in October of 1988 and contains 85 galaxies listed in the 2nd Reference Catalogue of Bright Galaxies.
plt.xticks(size=12)
plt.yticks(size=12)
plt.legend(edgecolor='black')
plt.show()
peak_wave=((const.c)/min(frequency))*u.angstrom #calculating the peak wavelength
print('Peak Wavelength: ' ,peak_wave.to(u.m),)
peak_freq=(max(frequency))/u.angstrom
print('Peak Frequency: ',peak_freq,)
WienTemp=(0.0029)/(peak_wave.to(u.m))
print('Temperature: ',WienTemp*u.K,)
```
